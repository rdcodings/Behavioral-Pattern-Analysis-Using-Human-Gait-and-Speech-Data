### 1. Load the Data

### 3. Visualize Distributions with Histograms
Histograms help us understand the distribution of individual features.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms
plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
sns.histplot(df['avg_step_length'], kde=True)
plt.title('Average Step Length Distribution')

plt.subplot(2, 3, 2)
sns.histplot(df['avg_stride_length'], kde=True)
plt.title('Average Stride Length Distribution')

plt.subplot(2, 3, 3)
sns.histplot(df['avg_walking_speed'], kde=True)
plt.title('Average Walking Speed Distribution')

plt.subplot(2, 3, 4)
sns.histplot(df['avg_acceleration'], kde=True)
plt.title('Average Acceleration Distribution')

plt.subplot(2, 3, 5)
sns.histplot(df['variance_step_length'], kde=True)
plt.title('Variance of Step Length Distribution')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plots
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.scatterplot(x=df['avg_step_length'], y=df['avg_walking_speed'])
plt.title('Step Length vs Walking Speed')

plt.subplot(2, 2, 2)
sns.scatterplot(x=df['avg_step_length'], y=df['avg_acceleration'])
plt.title('Step Length vs Acceleration')

plt.subplot(2, 2, 3)
sns.scatterplot(x=df['avg_step_length'], y=df['variance_step_length'])
plt.title('Step Length vs Variance of Step Length')

plt.subplot(2, 2, 4)
sns.scatterplot(x=df['avg_walking_speed'], y=df['avg_acceleration'])
plt.title('Walking Speed vs Acceleration')

plt.tight_layout()
plt.show()
```

The scatter plots visualize the relationships between different pairs of gait features. Here’s an explanation of each plot:

### 1. Step Length vs Walking Speed

- **Insight**: As the step length increases, the walking speed also tends to increase. This is expected, as longer steps generally lead to higher walking speeds.

### 2. Stride Length vs Walking Speed
- **Observation**: There is a strong positive correlation between average stride length and average walking speed.
- **Insight**: Similar to step length, as stride length increases, walking speed also increases. This is intuitive because a longer stride covers more distance per step, contributing to higher speeds.

### 3. Step Length vs Acceleration
- **Observation**: There is no clear correlation between average step length and average acceleration.
- **Insight**: The relationship between step length and acceleration appears random, indicating that step length alone may not be a strong predictor of acceleration.

### 4. Stride Length vs Acceleration
- **Observation**: There is no clear correlation between average stride length and average acceleration.
- **Insight**: Similar to step length, stride length does not show a strong relationship with acceleration, suggesting other factors may influence acceleration.

### 5. Step Length vs Variance of Step Length
- **Observation**: There is a negative correlation between average step length and the variance of step length.
- **Insight**: Higher average step lengths are associated with lower variance, indicating more consistent step lengths. Conversely, shorter steps tend to have higher variability.

### 6. Stride Length vs Variance of Step Length
- **Observation**: There is a negative correlation between average stride length and the variance of step length.
- **Insight**: Similar to step length, higher average stride lengths are associated with lower variance, suggesting that individuals with longer strides tend to have more consistent stride lengths.

### Overall Insights
- **Positive Correlations**: Both step length and stride length show positive correlations with walking speed.
- **No Clear Correlation**: Acceleration does not show a clear relationship with either step length or stride length.
- **Negative Correlations**: Both step length and stride length show negative correlations with the variance of step length, indicating that longer steps and strides are more consistent.

These scatter plots provide valuable insights into how different gait features relate to each other, helping to understand the underlying patterns in the data.

### 6. Correlation Heatmap
A correlation heatmap shows the correlation coefficients between features, helping us understand how they are related.

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Drop the 'avg_stride_length' column from the DataFrame before computing the correlation matrix
df_filtered = df.drop(columns=['avg_stride_length'])

# Correlation heatmap
plt.figure(figsize=(10, 8))
correlation_matrix = df_filtered.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

The correlation heatmap shows the correlation coefficients between the different features in your dataset. Here’s an explanation of the results:

#### Understanding Correlation
**Correlation Coefficient (r)**: This value ranges from -1 to 1.
+1: Perfect positive correlation; as one variable increases, the other also increases.
-1: Perfect negative correlation; as one variable increases, the other decreases.
0: No correlation; the variables do not have a linear relationship.

#### Analysis of the Correlation Heatmap

**avg_step_length**:

Has a perfect positive correlation (r = 1.00) with avg_stride_length, avg_walking_speed, and itself. This is expected because avg_stride_length and avg_walking_speed are derived from avg_step_length in your calculations.
Has a moderate positive correlation (r = 0.59) with variance_step_length, indicating that as the average step length increases, the variance in step length also tends to increase.
Has a very weak negative correlation (r = -0.06) with avg_acceleration, indicating almost no linear relationship.

**avg_stride_length**:

Has the same correlations as avg_step_length because it is directly derived from it.
Perfect positive correlation (r = 1.00) with avg_step_length, avg_walking_speed, and itself.
Moderate positive correlation (r = 0.59) with variance_step_length.
Very weak negative correlation (r = -0.06) with avg_acceleration.

**avg_walking_speed**:

Again, shows perfect positive correlation (r = 1.00) with avg_step_length, avg_stride_length, and itself.
Moderate positive correlation (r = 0.59) with variance_step_length.
Very weak negative correlation (r = -0.06) with avg_acceleration.

**avg_acceleration**:

Shows no significant linear relationship with avg_step_length, avg_stride_length, or avg_walking_speed (r = -0.06).
Weak negative correlation (r = -0.19) with variance_step_length, indicating a slight inverse relationship between acceleration and the variability of step length.

**variance_step_length**:

Moderate positive correlation (r = 0.59) with avg_step_length, avg_stride_length, and avg_walking_speed.
Weak negative correlation (r = -0.19) with avg_acceleration.

#### Insights
Perfect Correlation: The perfect correlation among avg_step_length, avg_stride_length, and avg_walking_speed indicates that these variables are linearly dependent on each other.
Variance Relationship: The moderate positive correlation between variance_step_length and the other step-related variables suggests that as the average values increase, there is also an increase in variability.
Acceleration: The weak correlations with avg_acceleration indicate that acceleration is relatively independent of the other measured features in your dataset.

#### Conclusion
This heatmap confirms that avg_step_length, avg_stride_length, and avg_walking_speed are highly interrelated. It also highlights that the variance in step length tends to increase with the average step and stride lengths. However, acceleration does not show strong linear relationships with the other features, suggesting it might be influenced by different factors or captured through non-linear dynamics.
